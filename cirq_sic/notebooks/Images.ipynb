```python
from cirq_sic import *
from cirq.contrib.svg import SVGCircuit
```

## Images

```python
n = 3
c = cirq.NamedQubit("c")
t = cirq.LineQubit.range(n)
circuit = cirq.Circuit(QCZ(c, t))
SVGCircuit(circuit)
```

```python
n = 3	
t = cirq.LineQubit.range(n)
c = cirq.LineQubit.range(n, 2*n)
circuit = cirq.Circuit(CZ(c, t), strategy=cirq.InsertStrategy.NEW_THEN_INLINE)
SVGCircuit(circuit)
```

```python
n = 2
t1 = cirq.LineQubit.range(n)
t2 = cirq.LineQubit.range(n, 2*n)
c = cirq.LineQubit.range(2*n, 3*n)
circuit = cirq.Circuit((CX(c, t1), qft(c, inverse=True), CX(c, t2), qft(c)), strategy=cirq.InsertStrategy.NEW_THEN_INLINE)
SVGCircuit(circuit)
```
